```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report
```

```python
train_data = pd.read_csv('/content/Titanic_Train.csv')
test_data=pd.read_csv('/content/Titanic_Test.csv')
```

```python
train_data.isnull().sum()
```

```python
train_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lr= LabelEncoder()
train_data['Sex']= lr.fit_transform(train_data['Sex'])
test_data['Sex']=lr.transform(test_data['Sex'])

train_data['Embarked']=lr.fit_transform(train_data['Embarked'])
test_data['Embarked']=lr.transform(test_data['Embarked'])
```

```python
train_data.fillna(0,inplace=True)
test_data.fillna(0,inplace=True)
```

```python
train_data.isnull().sum()
```

```python
test_data.isnull().sum()
```

```python
train_data.head(8)
```

```python
feactures=['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']
x_train=train_data[feactures]
y_train=train_data['Survived']
x_test=test_data[feactures]
```

```python

nb= GaussianNB()
nb.fit(x_train, y_train)
```

```python
y_train_pred = nb.predict(x_train)
y_test_pred = nb.predict(x_test)
accuracy = accuracy_score(y_train, y_train_pred)
print(f"Accuracy on the training dataset: {accuracy}")
```
