```python
import pandas as pd
df=pd.read_csv("/content/iris.csv")

import matplotlib.pyplot as plt
import seaborn as sns
iris=sns.load_dataset("iris")
iris.head()
#importing libraries for visualization
```

```python
plt.figure(figsize=(8,6))
sns.histplot(iris['sepal_length'])
plt.xlabel("sepal length(cm)")
plt.ylabel("Count")
plt.title("Histogram of sepal length")
plt.show
# for printing histogram, we use the histplot keyword for visualization. Inside keyword we give the dataset and specific feature that we wanted to plot in histogram
# from the graph we can see that sepal length between 5.5 and 6.25 has maximum count
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="variety",y="sepal.length",data=df)
plt.xlabel("species")
plt.ylabel("sepal length")
plt.title("Box Plot of Sepal Length")
plt.show
# for printing box plot, we use the keyword boxplot. Inside that we have to specify the parameters along with dataframe
# from the graph we can conclude that setosa has sepal length between 4.75 and 5.25, versicolor has sepal length between 5.6 and 6.25, virginica has sepal length between 6.4 and 6.75
```

```python
plt.figure(figsize=(8,6))
sns.kdeplot(df["sepal.width"],fill=True)
plt.xlabel('Sepal Width')
plt.ylabel('Count')
plt.title('Distribution Chart of sepal width')
plt.show()
# for printing distribution chart, we use the keyword kde plot. Inside key we've to specify the data frame and feature name
# from the graph we can conclude that sepal width 3 has highest count
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="sepal.length",y="sepal.width",hue="variety",data=df)
plt.xlabel("sepal length")
plt.ylabel("sepal width")
plt.title("Scatter plot of sepal length vs sepal width")
plt.show()
# for scatter plot printing, we use keyword scatter plot. we have to specify x,y coordinates(features) along with dataframe
# from the graph we can conclude that setosa has sepal length between 4 and 6, and sepal width between 3 and 4.5. versicolor has sepal length between 5 and 7 and sepal width between 2.25 and 3.25. virginica has sepal length between 5 and 8 and sepal width between 2.25 and 3.75
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="petal.length",y="petal.width",size="sepal.length",hue="variety",data=df)
plt.xlabel("petal length")
plt.ylabel("petal width")
plt.title("Scatter plot of petal length vs petal width")
plt.show()
# for printing scatter plot we use the keyword scatterplot. we have to specify the x,y coordinates
```

```python
sns.set(style="ticks")
sns.pairplot(df,hue="variety",markers=["o","s","d"])
plt.suptitle("Scatter matrix (pair plot)",y=1.02)
plt.show
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="sepal.length",y="sepal.width",hue="variety",size="petal.length",data=df)
plt.xlabel("sepal length")
plt.ylabel("sepal width")
plt.title("Bubble chart of sepal lenth vs sepal width(coloured by species,sized by petal length)")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.kdeplot(x=df["petal.length"],y=df["petal.width"],cmap="Blues",fill=True)
plt.xlabel("petal length")
plt.ylabel("petal width")
plt.title("Density chart of sepal length vs sepal width")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.set(style="whitegrid")
pd.plotting.parallel_coordinates(df,"variety",colormap="viridis")
plt.title("Parallel coordinates chart")
plt.show()
```

```python
features="sepal.length"
meanvalue=df[features].mean()
df["deviation"]=df[features]-meanvalue
plt.figure(figsize=(8,6))
plt.plot(df.index,df["deviation"],marker="o",linestyle="-",color='r')
plt.xlabel("Data Points")
plt.ylabel(f"fe Deviation from mean{features.capitalize()}")
plt.title(f"deviation chart for {features.capitalize()}")
plt.grid(True)
plt.show()

```

```python
plt.figure(figsize=(8,6))
sns.set(style="ticks")
pd.plotting.andrews_curves(df,"variety",colormap="viridis")
plt.title("Andrews Curve")
plt.show()
```
